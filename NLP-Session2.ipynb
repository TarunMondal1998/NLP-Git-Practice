## Text Cleaning

```python
import pandas as pd
import numpy as np
import re
import nltk
import openpyxl

import warnings
warnings.filterwarnings('ignore')
```

```python
data_cleaning = pd.read_excel("data_version2.xlsx")
```

```python
data_cleaning.head()
```

```python
data_cleaning = data_cleaning.drop(['Unnamed: 0'], axis = 1)
```

```python
data_cleaning.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string
import emoji
```

```python
data_cleaning['Review']= data_cleaning['ReviewHeader']+ ' ' + data_cleaning['ReviewBody']
data_version3 = data_cleaning [['Review', 'Review category']]
data_version3.head()
```

```python
print(data_version3['Review'][0])
```

```python
## Lowercasing the review
data_version3 ['Review'] = data_version3['Review'].str.lower()
```

```python
data_version3.head()
```

```python
## Removing HTML tags
def remove_html_tags (text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text) ## Function defination
data_version3['Review'] = data_version3['Review'].apply(remove_html_tags)
```

```python
data_version3.head()
```

```python
## Removing URLs
def remove_urls (text):
    pattern = re.compile (r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
data_version3 ['Review'] = data_version3['Review'].apply(remove_urls)
```

```python
data_version3.head()
```

```python
## Testing of remove_html_tags and remove_urls functions
sample_text = "https://media.gcflearnfree.org/global/coding/basketballdog.png I have a white cat which has two beautiful eyes"
print(remove_html_tags(sample_text))
print(remove_urls(sample_text))
```

```python
## Removing punctuations
import string
string.punctuation
punc = string.punctuation
def remove_punctuations (text):
    return text.translate(str.maketrans('', '',punc))
data_version3['Review']= data_version3['Review'].apply(remove_punctuations)
```
